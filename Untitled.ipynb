### Preprocessing

```python
import numpy as np
import itertools
import operator
```

```python
INPUT_PATH = 'seqdata.dat.txt'
MIN_SUPPORT = 0.01  # ex. 25% = 0.25
```

```python
list_cid_set = list()
file = open(INPUT_PATH, 'r')
for line_of_each_cid in file.readlines():
    each_trans_item_set = line_of_each_cid.split()[1:]
    list_cid_set.append(each_trans_item_set)
```

```python
list_cid_set[:10]
```

```python
from collections import defaultdict
list_cid_set_with_dict_tids = list()
item_set = set()
for each_cid_tids in list_cid_set:
#     print(each_cid_tids)
    tids = defaultdict(list)
    for idx, val in enumerate(each_cid_tids):
        if idx % 2 == 0:
            tids[val].append(each_cid_tids[idx+1])
            item_set.add(each_cid_tids[idx+1])
    list_cid_set_with_dict_tids.append(list(tids.values()))
```

```python
list_cid_set_with_dict_tids[:5]
```

```python
num_of_cid_set = len(list_cid_set_with_dict_tids)  ## total cid number
num_of_cid_set
```

```python
num_of_max_trans = max(len(x) for x in list_cid_set_with_dict_tids)  ## max trans length in cid
num_of_max_trans
```

```python
num_of_item_set = len(item_set)  ## total item number (random code)
num_of_item_set
```

### one-item

```python
npm_one_item_table = np.zeros(( num_of_cid_set * num_of_max_trans, num_of_item_set ))
```

```python
npm_one_item_table.shape  ## (260000, 3029) for current data
```

```python
item_set = list(map(int, item_set))
dict_item_idx = dict(map(lambda t: (t[1], t[0]), enumerate(sorted(item_set))))  ## assign new idx to each item for Matrix usage
```

```python
dict_item_idx
```

```python
dict_of_item_support = dict() ## using oringinal item id
for each_cid_idx, each_cid_val in enumerate(list_cid_set_with_dict_tids):
    set_of_item_support = set()
    for idx, val_set in enumerate(each_cid_val):
        lookup_idx_set = list()
        for val in val_set:
            lookup_idx_set.append(dict_item_idx[int(val)])
            set_of_item_support.add(int(val))
        npm_one_item_table[( each_cid_idx * num_of_max_trans ) + idx, [ int(x) for x in lookup_idx_set ]] = 1
    for uni_item in set_of_item_support:
        tmp_val = dict_of_item_support.get(uni_item, 0)
        dict_of_item_support[uni_item] = tmp_val + 1
```

```python
dict_of_item_support = dict(sorted(dict_of_item_support.items()))
dict_of_item_support
```

### DFS-Pruning
[ref.](https://blog.csdn.net/Kite_crazy/article/details/39735389)

```python
def dfs_pruning(item, Sn, In):
    print(item, Sn, In, 'start')
    global dict_item_idx
    global npm_one_item_table
    S_temp = list()
    I_temp = list()
    for s_candidate in Sn:
        print(item, s_candidate)
        result_column = get_associated_from_A2B_column(npm_one_item_table[:,item], npm_one_item_table[:,s_candidate])
        if (item, s_candidate) not in dict_item_idx:
            if check_single_column_match_min_support(result_column):
                S_temp.append(s_candidate)
                dict_item_idx[(item, s_candidate)] = max(dict_item_idx.items(), key=operator.itemgetter(1))[-1] + 1
                print(npm_one_item_table.shape, result_column.shape, 'pos1')
                npm_one_item_table = np.append(npm_one_item_table, result_column.reshape( num_of_cid_set * num_of_max_trans, 1), axis=1)
    for s_temp_candidate in S_temp:
        dfs_pruning(s_temp_candidate, S_temp, [x for x in S_temp if x > s_temp_candidate])
    for i_candidate in In:
        result_column = get_associated_from_AB_column(npm_one_item_table[:,item], npm_one_item_table[:,i_candidate])
        if check_single_column_match_min_support(result_column):
            I_temp.append(i_candidate)
            dict_item_idx[frozenset([item, i_candidate])] = max(dict_item_idx.items(), key=operator.itemgetter(1))[-1] + 1
            print(npm_one_item_table.shape, result_column.shape, 'pos2')
            npm_one_item_table = np.append(npm_one_item_table, result_column.reshape( num_of_cid_set * num_of_max_trans, 1), axis=1)
    for i_tmp_candidate in I_temp:
        dfs_pruning(i_tmp_candidate, S_temp, [x for x in I_temp if x > i_tmp_candidate])
        
```

```python
def scan_column_to_get_mask(column):
    list_temp_mask = list()
    for cid_idx in range(num_of_cid_set):
        for tran_idx in range(num_of_max_trans):
            if column[tran_idx + cid_idx * num_of_max_trans] == 1:
                list_temp_mask.append(0)
                for idx in range(tran_idx + 1, num_of_max_trans):
                    list_temp_mask.append(1)
                break
            else:
                list_temp_mask.append(0)
    return list_temp_mask
```

```python
def get_associated_from_A2B_column(a_column, b_column):
    a_column_mask = scan_column_to_get_mask(a_column)
    return np.logical_and(a_column_mask, b_column)
```

```python
def get_associated_from_AB_column(a_column, b_column):
    return np.logical_and(a_column, b_column)
```

```python
def check_single_column_match_min_support(column):
    support_count = 0
    for cid_idx in range(num_of_cid_set):
        for tran_idx in range(num_of_max_trans):
            if column[tran_idx + cid_idx * num_of_max_trans] == 1:
                support_count += 1
                break
    if support_count >= num_of_cid_set * MIN_SUPPORT:
        return True
    else:
        return False
```
