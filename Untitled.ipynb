### Preprocessing

```python
import numpy as np
```

```python
INPUT_PATH = 'seqdata.dat.txt'
MIN_SUPPORT = 0.01  # ex. 25% = 0.25
```

```python
list_cid_set = list()
file = open(INPUT_PATH, 'r')
for line_of_each_cid in file.readlines():
    each_trans_item_set = line_of_each_cid.split()[1:]
    list_cid_set.append(each_trans_item_set)
```

```python
# list_cid_set
```

```python
from collections import defaultdict
list_cid_set_with_dict_tids = list()
item_set = set()
for each_cid_tids in list_cid_set:
#     print(each_cid_tids)
    tids = defaultdict(list)
    for idx, val in enumerate(each_cid_tids):
        if idx % 2 == 0:
            tids[val].append(each_cid_tids[idx+1])
            item_set.add(each_cid_tids[idx+1])
    list_cid_set_with_dict_tids.append(list(tids.values()))
```

```python
num_of_cid_set = len(list_cid_set_with_dict_tids)  ## total cid number
num_of_cid_set
```

```python
num_of_max_trans = max(len(x) for x in list_cid_set_with_dict_tids)  ## max trans length in cid
num_of_max_trans
```

```python
num_of_item_set = len(item_set)  ## total item number (random code)
num_of_item_set
```

### one-item

```python
npm_one_item_table = np.zeros(( num_of_cid_set * num_of_max_trans, num_of_item_set ))
npa_one_item_support_array = np.zeros(num_of_item_set)
```

```python
dict_item_idx = dict(map(lambda t: (t[1], t[0]), enumerate(item_set)))  ## assign new idx to each item for Matrix usage
```

```python
dict_of_item_support = dict() ## using oringinal item id
for each_cid_idx, each_cid_val in enumerate(list_cid_set_with_dict_tids):
    set_of_item_support = set()
    for idx, val_set in enumerate(each_cid_val):
        lookup_idx_set = list()
        for val in val_set:
            lookup_idx_set.append(dict_item_idx[val])
            set_of_item_support.add(val)
        npm_one_item_table[( each_cid_idx * num_of_max_trans ) + idx, [ int(x) for x in lookup_idx_set ]] = 1
    for uni_item in set_of_item_support:
        tmp_val = dict_of_item_support.get(uni_item, 0)
        dict_of_item_support[uni_item] = tmp_val + 1
```

```python
def scan_column_to_get_mask(column):
    list_temp_mask = list()
    for cid_idx in range(num_of_cid_set):
        for tran_idx in range(num_of_max_trans):
            if column[tran_idx + cid_idx * num_of_max_trans] == 1:
                list_temp_mask.append(0)
                for idx in range(tran_idx + 1, num_of_max_trans):
                    list_temp_mask.append(1)
                break
            else:
                list_temp_mask.append(0)
    return list_temp_mask
```

```python
def get_associated_from_A2B_column(a_column, b_column):
    a_column_mask = scan_column_to_get_mask(a_column)
    return np.logical_and(a_column_mask, b_column)
```
